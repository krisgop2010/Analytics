```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

```python
data = pd.read_excel('C:\\DataSets\\Sales_Spend.xlsx')
data
```

```python
#calculate rolling mean
data_rm = data.rolling(3).mean()
```

```python
data_rm
```

```python
#select last 3 rows
data_rm1 = data_rm.iloc[6:9]
```

```python
data_rm1
```

```python
#take mean and convert to int..this will give us the next week's forecasted sales
data2 = data_rm1.mean().astype(int)
```

```python
data2
```

```python
# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter('C:\\DataSets\\week11forecast.xlsx', engine='xlsxwriter')
```

```python

data2.to_excel(writer, 'Sheet1')
```
